```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import requests
import StringIO
import zipfile
```

```python
url= 'http://seanlahman.com/files/database/lahman-csv_2014-02-14.zip'
r=requests.get(url).content
s=StringIO.StringIO(r)
zf = zipfile.ZipFile(s, 'r')
print zf.namelist()
```

```python
salary_csv = zf.open(zf.namelist()[zf.namelist().index('Salaries.csv')])
salaries = pd.read_csv(salary_csv)
```

```python
print "Number of rows: %i" %salaries.shape[0]
```

```python
salaries.head(10)
```

```python
team_csv = zf.open(zf.namelist()[zf.namelist().index('Teams.csv')])
teams = pd.read_csv(team_csv)
print "Number of rows %i" %teams.shape[0]
teams.head(10)
```

```python
total_salaries = salaries.groupby(['yearID','teamID'], as_index=False).sum()
```

```python
total_salaries.head()
```

```python
team_wins = teams[['yearID', 'teamID', 'W']].copy()
```

```python
team_wins.head()
```

```python
team_salary_wins = pd.merge(total_salaries, team_wins, how="inner", on=['yearID', 'teamID'])
```
